```python
import sys
print(sys.version_info) 
import numpy as np
import matplotlib.pyplot as plt 
import scipy.interpolate as interpolate
import nibabel as nib 
import os
```

```python
def nib_load(file_name):
    proxy = nib.load(file_name)
    data = proxy.get_data()
    proxy.uncache()
    return data
```

```python
path = '/home/ricardo/MyProgram/Task1/Brats18_dataset/MICCAI_BraTS_2018_Data_Training/HGG/Brats18_TCIA01_412_1/Brats18_TCIA01_412_1_flair.nii.gz'
```

```python
test = nib_load(path)
print(test.shape) 
plt.imshow(test[:,:,int(test.shape[2]/2)].T,cmap='gray',origin='lower')
plt.show()
```

```python
max_value = np.max(test)
min_value = np.min(test)

test = (test - min_value) / (max_value - min_value)
```

```python
test.shape
```

```python
test[np.where(test>0)]=1
```
